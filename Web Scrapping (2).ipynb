# Scrapping Data from a Website

```python
from bs4 import BeautifulSoup 
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_medical_colleges_in_India'

page = requests.get(url)

soup= BeautifulSoup(page.text,'html')
```

```python
soup.find_all('table');
```

```python
soup.find_all('table')[0];
```

```python
table = soup.find_all('table')[0]
```

```python
table.find_all('th');
```

```python
titles = table.find_all('th')

```

```python
col_titles = [title.text.strip() for title in titles]
```

```python
print(col_titles)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame (columns=col_titles)
df
```

```python
th_value = table.find_all('tr')
th_value;
```

```python
for i in th_value[1:]:
    row_data = i.find_all('td')
    table_rows=[data.text.strip() for data in row_data]
    
    
    length = len(df)
    df.loc[length] = table_rows
```
